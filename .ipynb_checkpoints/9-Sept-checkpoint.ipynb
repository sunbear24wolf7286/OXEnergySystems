```python
import math 
```

```python
def function(x):
    return math.exp(-x/3)*(2-x)-1
```

```python
def derivative(x):
    return math.exp(-x/4)*(-3/2 + x/4)
```

```python
tolerance = 0.00000001
guess = 5
f = function(guess)
df = derivative(guess)
deltax = - f/df
newguess = guess + deltax
error = math.fabs(deltax/newguess)
while error > tolerance:
    guess = newguess
    f = function(guess)
    df = derivative(guess)
    deltax = -f/df
    newguess = guess + deltax
    error = math.fabs(deltax/newguess)
    print(guess)
print("final answer = ",guess)
```

```python
print("answer = " + str(5))
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression 
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([2 + np.random.rand(1), 4 + np.random.rand(1), 6 + np.random.rand(1), 8 + np.random.rand(1)])
```

```python
LR = LinearRegression()
LR.fit(x.reshape(-1,1),y)
```

```python
LR.intercept_,LR.coef_
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
ytrend = np.array(x)*LR.coef_[0] + LR.intercept_
plt.plot(x,ytrend)
plt.text(1, 8, f"This is R^2 = {LR.score(x.reshape(-1,1), y):.3f}")
```
